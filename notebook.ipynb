```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("CardioGoodFitness.csv")
data.head()
```

```python
data.describe()
```

```python
means = data.mean()
data.fillna(value = means, inplace = True)
data.to_csv("CardioGoodFitness.csv")
```

```python
products = ['TM195', 'TM498', 'TM798']

for product in products:
    subset = data[data['Product'] == product]

    fig, axs = plt.subplots(3, 3, figsize=(12, 8))
    
    for i, col in enumerate(subset.columns[1:]):
        row, col_num = divmod(i, 3)
        axs[row, col_num].hist(subset[col], bins=20, edgecolor='black')
        axs[row, col_num].set_title(col)
    fig.suptitle(f'Histograms for {product}')
    plt.show()
```

```python
products = ['TM195', 'TM498', 'TM798']

for product in products:
    subset = data[data['Product'] == product]
    
    print(f'Profile for {product}:')
    print(f'Mean Age: {subset["Age"].mean()}')
    print(f'Median Age: {subset["Age"].median()}')
    print(f'Standard Deviation of Age: {subset["Age"].std()}')
    print(f'Mean Income: {subset["Income"].mean()}')
    print(f'Median Income: {subset["Income"].median()}')
    print(f'Standard Deviation of Income: {subset["Income"].std()}')
    print(f'Mean Miles: {subset["Miles"].mean()}')
    print(f'Median Miles: {subset["Miles"].median()}')
    print(f'Standard Deviation of Miles: {subset["Miles"].std()}')
    print('\n')
```
